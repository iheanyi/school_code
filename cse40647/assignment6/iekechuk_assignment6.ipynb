```python
from itertools import combinations

def load_products():
    products = dict()
    infile = open('bakery/products.csv', 'r')
    for line in infile:
        product_line = line.strip().strip(' ').split(',')
        products[product_line[0]] = product_line[1]
    
    return products

products = load_products()


def load_transactions(filename):
    T = []
    infile = open(filename, 'r')
    for line in infile:
        item = line.strip().replace(" ", "").split(',')[1:]
        T.append(item)
    
    return T

def format_itemset(_set):
    global products
    
    itemset_string = "[" + ", ".join(map(lambda x: products[x], _set)) + "]"
    return itemset_string

def generate_candidate_set(dataset):
    C = []
    for transaction in dataset:
        for item in transaction:
            if [item] not in C:
                C.append([item])
    C.sort()
    return map(frozenset, C)

def scanData(dataset, itemset, min_support=0.03):
    
    support_count = {}
    
    for transaction in dataset:
        for item in itemset:
            if item.issubset(transaction):
                support_count.setdefault(item, 0)
                support_count[item] += 1
    
    num_items = float(len(dataset))
    candidate_list = []
    support_data = {}

    for key, value in support_count.iteritems():        
        support = float(value/num_items)
        if support >= float(min_support):
            candidate_list.append(key)
            support_data.setdefault(key, 0)
            support_data[key] = support
        
    return candidate_list, support_data

def createJoinSet(itemset, length):
    return set([i.union(j) for i in itemset for j in itemset if len(i.union(j)) == length])

def print_itemset(current_cset, sup_data):
    for item in current_cset:
        print "Items: " + format_itemset(item) + "; Support: " + str(round(sup_data[item], 3))

def print_confidence(lhs, rhs, support_data, confidence):
    print format_itemset(lhs) + " =======> " + format_itemset(rhs)
    
def apriori(filename, minsupport=0.03, minconfidence=0.9):
    transactions = load_transactions(filename)
    C1 = generate_candidate_set(transactions)
    dataset = map(set, transactions)
    L1, support_data = scanData(dataset, C1, minsupport)

    c_set = {}
    k = 2
    current_lset = L1
    candidate_list = []
    
    print """ 
*************************************
    
        Frequent Itemset Numbers
    
*************************************
    
    """
    
    while(len(current_lset) != 0):
        c_set[k-1] = current_lset
        current_lset = createJoinSet(current_lset, k)
        current_cset, sup_data = scanData(dataset, current_lset)
        support_data.update(sup_data)
        print_itemset(current_cset, sup_data)
        current_lset = current_cset
        candidate_list.append(current_lset)
        k += 1
    print """ 
*************************************
    
        Assocation Rules
    
*************************************
    
    """   
    for item in candidate_list:
        association_rules(item, dataset, support_data, minconfidence)

def association_rules(candidate_set, dataset, support_data, minconfidence):    
    association_array = {}
    support_count = {}

    for item in candidate_set:
        for item_length in range(1, len(item)):
            for combo in combinations(item, item_length):
                lhs = frozenset(combo)
                rhs = item - lhs
                for transaction in dataset:
                    if lhs.issubset(transaction):# and rhs.issubset(transaction):
                        support_count.setdefault(lhs, 0)
                        support_count[lhs] += 1  
                
                num_items = float(len(dataset))
                support_s = float(support_count[lhs]) / num_items
                support_f = float(support_data[item])
                confidence = float(support_f / support_s)
                
                if confidence >= minconfidence:
                    print format_itemset(lhs) + " =======> " + format_itemset(rhs) + "\nSupport: " + str(round(support_s, 3)) + "; Confidence: " + str(round(confidence, 3)) + "\n"

apriori('bakery/1000/1000-out1.csv', 0.03, 0.9)
```
