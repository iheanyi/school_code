# Iris Dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

""" Attribute Information:
1. sepal length in cm 
2. sepal width in cm 
3. petal length in cm 
4. petal width in cm 
5. class: 
-- Iris Setosa 
-- Iris Versicolour 
-- Iris Virginica
"""

attributes_row = ["sepal length (cm)", "sepal width (cm)", "petal length (cm)", "petal width (cm)", "class"]

df = pd.read_csv('iris.data', header=None,names=attributes_row, usecols = range(4))

# Print first five entries.
print df.head(5)

print "\n"
print "Summary Statistics\n"
print df.describe()

print "Range"
print "-"*50

for key in df.keys():
    #if not isinstance(df[key].max(), str):
    print key + "\t" + str(df[key].max()-df[key].min())

plt.figure(figsize(16,16))
df.hist()

plt.figure()
df.boxplot()



```

#Conceptual Questions

##Iris Dataset

###Histograms
For the sepal length and sepal width, it is easy to visualize the general median values, in addition to the general bell-shape of the values, yet in the petal length and petal width histograms, there is no clearly visible median value, in addition to the data not being as presented neatly.

For the petal length histogram, it would probably be best to segment the histogram at the value 3, this would give a better visualization of the data.


###Box Plots
For the box plots, the medians tend to differ significantly between the sepal length and width features. The petal width and length features seem to have medians in fairly similar positions, located towards the upper portion of their box plots.

The petal length seems to explain the greatest amount of data. 

## Pen Digits Dataset
For the Digits dataset, I noticed numerous outliers present in the second (feature 1) and fourth (feature 3) features of the box plot, all below the minimum elements. When examining their histograms, one can see a lack of values from 0-20
, whereas other features have fair amount of data points in this range. The second and fourth features have an unusual overload of higher values.
